```python
import pandas as pd
import numpy as np

df = pd.read_csv('./tested.csv')
df.head(5)
df.dropna(inplace=True)
```

##Use ‘Fare’ column and perform following transformations
####Logarithmic
####Reciprocal
####Square root
####Exponential
####Box-Cox
####Quantile


```python
#Logarithmic
fare_column = df['Fare']
fare_logarithmic = np.log1p(fare_column)
result_logarithmic = pd.DataFrame({'Fare_Logarithmic': fare_logarithmic})
result_logarithmic
```

```python
#Reciprocal
fare_reciprocal = 1 / fare_column
result_reciprocal = pd.DataFrame({'Fare_Reciprocal': fare_reciprocal})
result_reciprocal
```

```python
#Square Root
fare_square_root = np.sqrt(fare_column)
result_square_root = pd.DataFrame({'Fare_SquareRoot': fare_square_root})
result_square_root
```

```python
#Exponential
fare_exponential = np.exp(fare_column)
result_exponential = pd.DataFrame({'Fare_Exponential': fare_exponential})
result_exponential
```

```python
#Box-Cox
from scipy.stats import boxcox
fare_column_positive = fare_column + abs(fare_column.min()) + 1e-6
fare_boxcox, _ = boxcox(fare_column_positive)
result_boxcox = pd.DataFrame({'Fare_BoxCox': fare_boxcox})
result_boxcox
```

```python
#Quantile
from sklearn.preprocessing import QuantileTransformer
quantile_transformer = QuantileTransformer(output_distribution='uniform')
fare_quantile = quantile_transformer.fit_transform(fare_column.values.reshape(-1, 1))
result_quantile = pd.DataFrame({'Fare_Quantile': fare_quantile.flatten()})
result_quantile
```
