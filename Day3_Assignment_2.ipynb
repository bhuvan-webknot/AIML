```python
import pandas as pd
import numpy as np

df = pd.read_csv("./tested.csv")
df.head()
```

###Perform one-hot encoding: replace the categorical variable by different boolean variables (0/1) to indicate whether or not certain label is true for that observation

```python
one_hot_encoded = pd.get_dummies(df, columns=['Sex','Pclass'])
one_hot_encoded
```

###Perform ordinal - encoding: replace the labels by some ordinal number if ordinal is meaningful


```python
categorical_columns = ['Sex','Pclass']
df1=df.copy(True)
for col in categorical_columns:
    df1[col + '_ordinal_encoded'] = df[col].astype('category').cat.codes
df1
```

###Perform Mean encoding: replace the label by the mean of the target for that label (the target must be 0/1 valued or continuous)


```python
df2=df1.copy(True)
categorical_col = ['Sex', 'Pclass']

for col in categorical_col:
    mean_encoded = df2.groupby(col)['Survived'].mean()
    df2[col + '_mean_encoded'] = df2[col].map(mean_encoded)
df2.head()
```

###Perform target-encoding: similar to mean encoding, but use both posterior probability and probability of the target

```python
df3=df2.copy(True)
categorical_cols = ['Sex', 'Pclass']

target_encoding_mapping = {}
for column in categorical_cols:
    target_encoding_mapping[column] = df3.groupby(column)['Survived'].transform('mean')

# Apply target encoding to each categorical column
for column in categorical_columns:
    df3[column+'_target_encoded'] = target_encoding_mapping[column]

df3.head()
```

###Perform Weight of Evidence (WOE) encoding: replace the label with WOE of each label. WOE is computed from the basics odds ratio [ In((Proportion of Good Outcomes)/ (Proportion of Bad Outcomes) ]

```python
df4=df3.copy(True)
categorical_cols = ['Sex', 'Pclass']

for column in categorical_columns:
  good_outcomes_proportion = df4.groupby(column)['Survived'].mean()
  bad_outcomes_proportion = 1 - good_outcomes_proportion
  woe = np.log(good_outcomes_proportion / bad_outcomes_proportion)
  df4[column+'woe_encoded'] = df4['Survived'].map(woe)

df4
```
