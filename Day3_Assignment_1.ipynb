```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./vgsales.csv")
df.head()
```

```python
#Check the total number and percentage of missing values per column in the dataset
missing_values = df.isnull().sum()
percentage = (missing_values/len(df))*100
percentage
```

```python
#Exclude all cases (listwise) with missing values
df1 = df.copy(True)
df.dropna(inplace = True)
df1.isnull().sum()
```

```python
#Create an additional column ‘NA’ indicating whether the data was missing for that observation
df1["NA"] = 0
df1["NA"] = df1.isnull().any(axis=1).astype(int)
df1[(df1["NA"]==1)]

#Arbitrary Value Imputation - replace the NA by arbitrary values
df2 = df1.copy(True)
df1.fillna(999,inplace=True)
df1[(df1["NA"]==1)]
```

```python
#Mean/Median/Mode Imputation - replace the NA be mean/median/mode of that column
df3 = df2.copy(True)
for column in ['Year','Publisher']:
    df2[column].fillna(df2[column].mode()[0], inplace=True)
df2[(df2["NA"]==1)]
```

```python
#End of distribution Imputation - replace the NA by values that are at the far end of the distribution of that column
df4 = df3.copy(True)
std = df3.std()
eod = df3.mean()+3*std
df3.fillna(eod,inplace=True)
df3[df3["NA"]==1]
```

```python
#Random Imputation - Replace the NA with random sampling from the pool of available observation of the column
df5 = df4.copy(True)
def random_imputation(column):
    non_na_values = column.dropna()
    na_count = column.isna().sum()
    if na_count == 0:
        return column
    random_samples = np.random.choice(non_na_values, na_count)
    column[column.isna()] = random_samples
    return column

data_imputed = df4.apply(random_imputation)
df4[df4["NA"]==1]
```

```python
#Identify outliers based on arbitrary boundaries
df6 = df5.copy(True)
numerical_col = df5.select_dtypes(include=['float'])
lower=0
upper=20
numerical_col["Year"]=1
outliers = numerical_col[(numerical_col<lower) | (numerical_col>upper)].any(axis=1)
outliers
```

```python
#Identify outliers by IQR - Interquartile Range Rule
df7 = df6.copy(True)
def identify_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (column < lower_bound) | (column > upper_bound)
    return outliers

outliers_mask = numerical_col.apply(identify_iqr).any(axis=1)
outliers_mask
```

```python
#Identify outliers by Mean and Standard Deviation
def identify_outliers_mean_std(column):
    mean = column.mean()
    std = column.std()
    lower_bound = mean - 3 * std
    upper_bound = mean + 3 * std
    outliers = (column < lower_bound) | (column > upper_bound)
    return outliers

outliers_mask = numerical_col.apply(identify_outliers_mean_std).any(axis=1)
outliers_mask
```

```python
#Identify outliers by Median and Median Absolute Deviation
def identify_outliers_median_mad(column):
    median = column.median()
    mad = np.median(np.abs(column - median))

    # Set a threshold for outliers (e.g., 3 times MAD)
    threshold = 3
    lower_bound = median - threshold * mad
    upper_bound = median + threshold * mad
    outliers = (column < lower_bound) | (column > upper_bound)
    return outliers

outliers_mask = numerical_col.apply(identify_outliers_median_mad).any(axis=1)
outliers_mask
```

```python
from scipy.stats.mstats import winsorize
```

```python
#Winsorization - Top coding and bottom coding (capping the maximum of a distribution at an arbitrary set value, vice versa)

def winsorize_columns(data, columns):
    limits=(0.05, 0.05)
    for column in columns:
        data[column] = winsorize(data[column],limits, inplace=True)
winsorize_columns(numerical_col, numerical_col)
```

```python
#Discard outliers
from scipy import stats
z_score_threshold = 3

# Identify outliers using Z-score
z_scores = np.abs(stats.zscore(df6.select_dtypes(include='number')))
outlier_mask = (z_scores > z_score_threshold).any(axis=1)
data_no_outliers = numerical_col[~outlier_mask]
data_no_outliers
```

```python
#Replace the outlier by mean/median/most frequent values of that column
numeric_cols = df2.select_dtypes(include=np.number)
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))).any(axis=1)
new_df=numeric_cols[outliers]
new_df['outliers']=outliers

for col in new_df:
  if new_df[col].dtype!='object':
    if new_df[col].dtype == 'float' or new_df[col].dtype == 'int':
        replacement_value = new_df[col].mean()
    new_df.loc[outliers, col] = replacement_value
new_df
```

```python
#Group the observations that show rare labels into a unique category
df7 = df6.copy(True)
def group_rare_labels(column, threshold=0.05):
    value_counts = column.value_counts(normalize=True)
    rare_labels = value_counts[value_counts < threshold].index
    column = column.apply(lambda x: 'Other' if x in rare_labels else x)
    return column

categorical_columns = df7.select_dtypes(include='object')
data_grouped_rare_labels = categorical_columns.apply(group_rare_labels)
data_grouped_rare_labels
```

```python
#Mode Imputation - replace the rare label by most frequent label
def mode_imputation(column, threshold=0.05):
    value_counts = column.value_counts(normalize=True)
    rare_labels = value_counts[value_counts < threshold].index
    mode_val = column.mode().iloc[0]
    column = column.apply(lambda x: mode_val if x in rare_labels else x)
    return column

categorical_columns = df7.select_dtypes(include='object')
data_mode_imputed = categorical_columns.apply(mode_imputation)
data_mode_imputed
```
