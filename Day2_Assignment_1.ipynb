#Visualization

```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./tested.csv")
df.head()
```

```python
#Get the dtypes for each columns and  general data description of the entire dataset
df.dtypes
df.describe()
```

```python
#Draw the barplot of a discrete variable x=pclass against target variable y=survived. What is the default value shown wrt to y?
survived_df = df[df['Survived'] == 1]
survived_df['Pclass'].value_counts().plot(kind='bar')
```

```python
#Draw the boxplot of a discrete variable x=pclass against y=fare
sns.boxplot(x='Pclass', y='Fare', data=df)
```

```python
#Draw the distplot of a continuous variable x=fare
sns.distplot(df['Fare'], bins=10)
```

```python
#Draw the scatter plot of two variables, x=fare and y=pclass
plt.scatter(df["Fare"],df["Pclass"])
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```
